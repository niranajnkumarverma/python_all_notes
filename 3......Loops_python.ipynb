## Difference between for and while loop?
- Number of iteration in for loop we knew.
- Number of iteration in while loop we didn't knew.
- Time taken by for loop lesser then while loop because while loop fully depanded on a condition.

```python
# basic * star pattern with while loop

i = 1
length = 11
while i < length:
  print('* ' * i)
  i += 1
```

```python
i = 11
length = 0
while i > length:
  print('* ' * i)
  i -= 1
```

```python
i = 1
length = 14
while i < length:
  print('* ' * (length - i) )
  i += 1
```

```python
i = 1
length = 11
while i < length:
  print('* ' * (length - i) * i )
  i += 1
```

```python
i = 1
length = 11
while i < length:
  print(' ' * (length - i) + '* ' * i )
  i += 1
```

```python
# make a list with unique elements using loop

names = ['jai', 'jaini', 'jai', 'neel', 'niranjan', 'kapil', 'vinay', 'jaini']

t = []
for i in names:
  if i not in t:
    t.append(i)
names = t
del t

print(names)
```

```python
# make a list with unique elements using set

names = ['jai', 'jaini', 'jai', 'neel', 'niranjan', 'kapil', 'vinay', 'jaini']

names = list( set(names) )

print(names)
```

```python
# Nested loop
l = [
     [1,2,3,4],
     ['a','b','c','d']
    ]

for i in l:
  for j in i:
    print(j)
```

# Control Statements

### 1. break
### 2. continue
### 3. pass

```python
for i in range(10):
  if i == 5:
    break
  print(i)
```

```python
for i in range(10):
  if i == 5:
    continue
  print(i)
```

```python
for i in range(5):
```

```python
for i in range(5):
  pass
```

```python
for i in range(5):
  if i % 2 == 0:
    print(i)
  else:
    pass
```
